## Prepare Data for the Classification Model



I need to import the required libraries, define the feature and target variables, encode the categorical target into numerical form, and then split the data into training and testing sets as per the subtask instructions. Finally, I will print the shapes of the split datasets to verify the operation.



```python
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split

# 2. Define la característica X como la columna xg_difference de matches_df.
# Asegúrate de que X sea un DataFrame o una Serie de pandas con la forma correcta para el modelo.
X = matches_df[['xg_difference']]

# 3. Define la variable objetivo y como la columna Match_Result de matches_df.
y = matches_df['Match_Result']

# 4. Inicializa LabelEncoder y ajusta y transforma la variable objetivo y
# para convertir las categorías de texto en valores numéricos.
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)

# Mostrar el mapeo de las etiquetas codificadas
print("Mapeo de las etiquetas codificadas:")
for i, category in enumerate(label_encoder.classes_):
    print(f"{category}: {i}")

# 5. Divide X y y_encoded en conjuntos de entrenamiento y prueba.
# Usar un test_size del 20-30% y un random_state para asegurar la reproducibilidad.
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.25, random_state=42)

# 6. Imprime las formas de los conjuntos de entrenamiento y prueba para verificar la división.
print(f"\nShape de X_train: {X_train.shape}")
print(f"Shape de X_test: {X_test.shape}")
print(f"Shape de y_train: {y_train.shape}")
print(f"Shape de y_test: {y_test.shape}")
```

The previous step successfully prepared the data for model training. Now, I need to train a Logistic Regression model using the prepared training data (`X_train`, `y_train`).



```python
from sklearn.linear_model import LogisticRegression

# 1. Inicializa el modelo de Regresión Logística.
# Dado que la variable objetivo tiene múltiples clases, se utiliza un clasificador multi-clase.
# 'multi_class' puede ser 'auto', 'ovr' (one-vs-rest) o 'multinomial'. 'auto' es un buen punto de partida.
# 'solver' es el algoritmo a usar para la optimización. 'liblinear' es bueno para datasets pequeños.
model = LogisticRegression(multi_class='multinomial', solver='lbfgs', random_state=42)

# 2. Entrena el modelo utilizando los conjuntos de entrenamiento.
print("Entrenando el modelo de Regresión Logística...")
model.fit(X_train, y_train)

print("Modelo de Regresión Logística entrenado exitosamente.")
```

The previous code block successfully trained the Logistic Regression model but produced a `FutureWarning` regarding the `multi_class` parameter. To address this and prevent future warnings, I will remove the explicitly set `multi_class='multinomial'` as `multinomial` is becoming the default for this solver. After training, the next logical step according to the overall task is to evaluate the model's performance on the test set.



```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
import numpy as np

# 1. Inicializa el modelo de Regresión Logística.
# Dado que la variable objetivo tiene múltiples clases, se utiliza un clasificador multi-clase.
# 'multi_class' puede ser 'auto', 'ovr' (one-vs-rest) o 'multinomial'. 'auto' es un buen punto de partida.
# 'solver' es el algoritmo a usar para la optimización. 'lbfgs' es adecuado para problemas multiclase.
# Removed multi_class='multinomial' to address FutureWarning, as 'lbfgs' solver will default to 'multinomial' for multiclass problems.
model = LogisticRegression(solver='lbfgs', random_state=42)

# 2. Entrena el modelo utilizando los conjuntos de entrenamiento.
print("Entrenando el modelo de Regresión Logística...")
model.fit(X_train, y_train)

print("Modelo de Regresión Logística entrenado exitosamente.")

# 3. Realizar predicciones en el conjunto de prueba.
y_pred = model.predict(X_test)

# 4. Evaluar el rendimiento del modelo.
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred, average='weighted', zero_division=0)
recall = recall_score(y_test, y_pred, average='weighted', zero_division=0)
f1 = f1_score(y_test, y_pred, average='weighted', zero_division=0)
conf_matrix = confusion_matrix(y_test, y_pred)

print(f"\nAccuracy: {accuracy:.2f}")
print(f"Precision: {precision:.2f}")
print(f"Recall: {recall:.2f}")
print(f"F1-score: {f1:.2f}")
print("Matriz de Confusión:\n", conf_matrix)

# To decode the numerical predictions back to original labels for better interpretation if needed.
# print("Original labels corresponding to encoded values:")
# for i, category in enumerate(label_encoder.classes_):
#     print(f"{i}: {category}")
```

## Preparar Nuevas Características del Ranking FIFA



I need to prepare new features by incorporating FIFA rankings for home and away teams into the `matches_df`. This involves creating a ranking mapping, performing two left merges with `matches_df` (one for home teams and one for away teams), renaming the new rank columns appropriately, and then displaying the updated DataFrame to verify the changes.



```python
import pandas as pd

# 1. Selecciona las columnas 'team' y 'rank' del DataFrame ranking_df
# para crear un DataFrame de mapeo de rankings.
# Renombra la columna 'rank' a 'team_rank' en este nuevo DataFrame.
ranking_map_df = ranking_df[['team', 'rank']].copy()
ranking_map_df.rename(columns={'rank': 'team_rank'}, inplace=True)

# 2. Fusiona matches_df con el DataFrame de mapeo de rankings para home_team.
# Renombra la columna resultante 'team_rank' a 'home_team_rank'.
matches_df = pd.merge(
    matches_df,
    ranking_map_df,
    left_on='home_team',
    right_on='team',
    how='left',
    suffixes=('', '_home_rank')
)
matches_df.rename(columns={'team_rank': 'home_team_rank'}, inplace=True)
matches_df.drop(columns=['team'], inplace=True) # Drop the redundant 'team' column from the merge

# 3. Realiza una segunda fusión de matches_df con el mismo DataFrame de mapeo de rankings para away_team.
# Renombra la columna resultante 'team_rank' a 'away_team_rank'.
matches_df = pd.merge(
    matches_df,
    ranking_map_df,
    left_on='away_team',
    right_on='team',
    how='left',
    suffixes=('', '_away_rank')
)
matches_df.rename(columns={'team_rank': 'away_team_rank'}, inplace=True)
matches_df.drop(columns=['team'], inplace=True) # Drop the redundant 'team' column from the merge

# 4. Muestra las primeras filas de matches_df con las nuevas columnas home_team_rank y away_team_rank
print("Matches DataFrame with new 'home_team_rank' and 'away_team_rank' columns:")
display(matches_df[['home_team', 'home_team_rank', 'away_team', 'away_team_rank', 'xg_difference', 'goal_difference']].head())
```

## Manejo de Valores Faltantes en Nuevas Características

First, I will inspect the `home_team_rank` and `away_team_rank` columns for null values. Then, I will impute any missing values in these columns with their respective means to prepare the data for modeling, and finally, I will verify that no nulls remain.



```python
import numpy as np

# 1. Inspeccionar la cantidad de valores nulos en 'home_team_rank' y 'away_team_rank'
print("Valores nulos antes de la imputación:")
print(matches_df[['home_team_rank', 'away_team_rank']].isnull().sum())

# 2. Imputar los valores nulos con la media de sus respectivas columnas
# Calcular la media de cada columna, ignorando NaNs
mean_home_rank = matches_df['home_team_rank'].mean()
mean_away_rank = matches_df['away_team_rank'].mean()

# Imputar los valores nulos
matches_df['home_team_rank'].fillna(mean_home_rank, inplace=True)
matches_df['away_team_rank'].fillna(mean_away_rank, inplace=True)

# 3. Verificar que no queden valores nulos
print("\nValores nulos después de la imputación:")
print(matches_df[['home_team_rank', 'away_team_rank']].isnull().sum())

# Mostrar las primeras filas del DataFrame con las columnas imputadas para verificación visual
print("\nPrimeras filas del matches_df con los rankings imputados:")
display(matches_df[['home_team', 'home_team_rank', 'away_team', 'away_team_rank']].head())
```

The previous code produced a `FutureWarning` due to using `inplace=True` with chained assignment. To address this and adhere to best practices for future pandas versions, I will modify the `.fillna()` calls to assign the results directly back to the respective columns.



```python
import numpy as np

# 1. Inspeccionar la cantidad de valores nulos en 'home_team_rank' y 'away_team_rank'
print("Valores nulos antes de la imputación:")
print(matches_df[['home_team_rank', 'away_team_rank']].isnull().sum())

# 2. Imputar los valores nulos con la media de sus respectivas columnas
# Calcular la media de cada columna, ignorando NaNs
mean_home_rank = matches_df['home_team_rank'].mean()
mean_away_rank = matches_df['away_team_rank'].mean()

# Imputar los valores nulos, assigning directly to avoid FutureWarning
matches_df['home_team_rank'] = matches_df['home_team_rank'].fillna(mean_home_rank)
matches_df['away_team_rank'] = matches_df['away_team_rank'].fillna(mean_away_rank)

# 3. Verificar que no queden valores nulos
print("\nValores nulos después de la imputación:")
print(matches_df[['home_team_rank', 'away_team_rank']].isnull().sum())

# Mostrar las primeras filas del DataFrame con las columnas imputadas para verificación visual
print("\nPrimeras filas del matches_df con los rankings imputados:")
display(matches_df[['home_team', 'home_team_rank', 'away_team', 'away_team_rank']].head())
```
